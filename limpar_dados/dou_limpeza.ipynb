```python
import bs4
import pandas as pd
import re
import requests

from unidecode import unidecode
```

```python
diario = pd.read_json("captura")
```

### Filtrando assuntos de interesse

```python
df = diario["paragraphs"]
```

```python
df
```

```python
# Separando os Atos de acordo com a temática de interesse - Indígena
indigena = []
for p in df:
    filtro = re.search(r"indígena", p) or re.search(r"índio", p)
    if filtro:
        indigena.append(p)
        print("Início: ", filtro.start())
        print("Fim: ", filtro.end())
        print(p)
```

```python
# Separando os Atos de acordo com a temática de interesse - Mineraão
mineracao = []
for p in df:
    filtro = re.search(r"mineração", p)
    if filtro:
        mineracao.append(p)
        print("Início: ", filtro.start())
        print("Fim: ", filtro.end())
        print(p)
        
```

### Tratamento

```python
# Limpar acentuação

sem_acento_ind = [unidecode(palavra) for palavra in indigena]
sem_acento_minera = [unidecode(palavra) for palavra in mineracao]
```

```python
# Limpar os espaços e criar uma lista de strings

def get_espacos(df):
    atos = []
    for p in range(len(df)):
        results = re.split(r"\\n", df)
        atos.append(results)
        return atos
```

```python
ind = get_espacos(str(sem_acento_ind))
minera = get_espacos(str(sem_acento_minera))
```

### Buscando no Google

```python
# Encontrar termos de interesse no Google:

texto = "Pirititi"

url = "https://google.com/search?q=" + texto 
request_result=requests.get(url) 
soup = bs4.BeautifulSoup(request_result.text, "html.parser") 
```

```python
# Pegar título e URL

lista = soup.find_all("div", {"class": "kCrYT"})

for e in lista:
    texto = e.find("h3")
    if texto is None:
        continue
    else:
        print("------------------")
        titulo = texto.getText()
        print(titulo)
        
        links = e.find("a", href=re.compile(r"[/]([a-z]|[A-Z])\w+"))
        link = str(links).split("url?q=")
        link = link[1]
        link = str(link).split('&amp;sa=U&amp')
        print(link[0])
```
